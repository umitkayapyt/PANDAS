```python
import pandas as pd ; import numpy as np
```

```python
v1=pd.DataFrame({"anahtar":["a","b","c","c","d","e"],"say1":range(6)})
v2=pd.DataFrame({"anahtar":["b","c","d","e"],"say2":range(4)})
```

```python
v1
```

```python
v2
```

```python
pd.merge(v1,v2)
```

```python
pd.merge(v1,v2, on="anahtar")
```

```python
v3=pd.DataFrame({"anahtar1":["a","b","c","c","d","e"],"say1":range(6)})
v4=pd.DataFrame({"anahtar2":["b","c","d","e"],"say2":range(4)})
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2")
```

```python
v3
```

```python
v4
```

```python
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2", how="outer")
```

```python
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2", how="left")
```

```python
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2", how="right")
```

```python
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2", how="inner")
```

```python
v4=pd.DataFrame({"anahtar":["a","b","c","c","d","e"],"say1":range(6),
                "rakam":["bir","uc","iki","bir","bir","iki"]})
v5=pd.DataFrame({"anahtar":["b","c","d","f"],"say2":range(4),
                "rakam":["bir","bir","iki","iki"]})
```

```python
v4
```

```python
v5
```

```python
pd.merge(v4,v5,on=["anahtar","rakam"],how="outer")
```

```python
pd.merge(v4,v5,on="anahtar",how="outer")
```

```python
pd.merge(v4,v5,on="anahtar",how="outer",suffixes=("_veri1","_veri2"))
```
