```python
vs=pd.DataFrame({"anahtar1":list("aabbab"),
                "anahtar2":["bir","iki","üç"]*2,
                "veri1":np.random.randn(6),
                "veri2":np.random.randn(6)})
vs
```

```python
grup=vs["veri1"].groupby(vs["anahtar1"]).mean()
grup
```

```python
ort=vs["veri1"].groupby([vs["anahtar1"],vs["anahtar2"]]).mean()
ort
```

```python
ort.unstack()
```

```python
vs.groupby("anahtar1").mean()
```

```python
vs.groupby(["anahtar1","anahtar2"]).mean()
```

```python
for isim, s in vs.groupby("anahtar2"):
    print(isim)
    print(s)
```

```python
for (x1,x2), s in vs.groupby(["anahtar1","anahtar2"]):
    print(x1,x2)
    print(s)
```
